```python
# Cosine similarity function
def get_similar_articles(q, df):
  print("query:", q)
  print("Here are the articles that are similar: ")
  # Convert the query become a vector
  q = [q]
  q_vec = vectorizer.transform(q).toarray().reshape(df.shape[0],)
  sim = {}
  # Calculate the similarity
  for i in range(10):
    sim[i] = np.dot(df.loc[:, i].values, q_vec) / np.linalg.norm(df.loc[:, i]) * np.linalg.norm(q_vec)
  
  # Sort the values 
  sim_sorted = sorted(sim.items(), key=lambda x: x[1], reverse=True)
  # Print the articles and their similarity values
  for k, v in sim_sorted:
    if v != 0.0:
      print("Similar values:", v)
      #print(clean_doclist[k])
      for d in documents_dictlist:
        if d["contents"] == clean_doclist[k]:
          print(d["pub_name"])

testSoup = BeautifulSoup(requests.get("https://pubmed.ncbi.nlm.nih.gov/?term=covid+19&size=20").content, 'html5lib')
urlResults = testSoup.findAll('a', attrs = {'class': 'docsum-title'})
documents = []
stop_words = set(stopwords.words('english'))

# NEW VERSION for DOCUMENTS
# keys: content, url, publication name
dictDocuments = {"contents": None, "url": None, "pub_name": None}
documents_dictlist = []

# TO DO: Add documents to a dictionary to find the URL so we can find citations/publication name/etc later.
# CONTINUE: function get_similar_articles uses k in sim sorted as the key of the clean doclist. You can use the dictionary and match based on the document
# 15.5 seconds
for el in urlResults:
    content = []
    url = "https://pubmed.ncbi.nlm.nih.gov/%s/"%(el["href"].replace("/",""))
    newSoup = BeautifulSoup(requests.get(url).content, 'html5lib').find('div', {'class': 'abstract-content selected'})
    title = BeautifulSoup(requests.get(url).content, 'html5lib').find('h1', {'class': 'heading-title'}).text
    # If the article has no abstract, find the full text link
    if newSoup is None:
      fullTextListLinks = BeautifulSoup(requests.get(url).content, 'html5lib').find('div', attrs = {'class':'full-text-links-list'})
      # Get the first full text link, does not matter which one
      fullTextListLinks= fullTextListLinks.findAll('a', attrs = {'class':'link-item'})[0]
      url = fullTextListLinks["href"]
      newSoup = BeautifulSoup(requests.get(url).content, 'html5lib')
    
    try:
        for y in newSoup.findAll("p"):
          # Add the text inside the paragraph into content
          content.append(y.text)
    except AttributeError as e:
        print("Error on %s"%(el["href"]))
    
    # New
    dictDocuments.update({"contents": ' '.join(content), "url": url, "pub_name": title})
    documents_dictlist.append(dictDocuments)
    
    # Old
    documents.append(' '.join(content))

# Clean documents
clean_doclist = []
for s in documents:
  # Remove Unicode
  doc = re.sub(r'[^\x00-\x7F]+', ' ', s)
  # Remove Mentions
  doc = re.sub(r'@\w+', '', doc)
  # Lowercase the document
  doc = doc.lower()
  # Remove punctuations
  doc = re.sub(r'[%s]' % re.escape(string.punctuation), ' ', doc)
  # Remove the numbers
  # Could be a problem in the future
  doc = re.sub(r'[0-9]', '', doc)
  # Remove the doubled space
  doc = re.sub(r'\s{2,}', ' ', doc)
  # Remove the stop words from each word in a document
  filtered_doc = ' '.join([word for word in doc.split(" ") if not word in stop_words])
  clean_doclist.append(filtered_doc)

for d in documents_dictlist:
  # Remove Unicode
  doc_dict = re.sub(r'[^\x00-\x7F]+', ' ', d["contents"])
  # Remove Mentions
  doc_dict = re.sub(r'@\w+', '', doc_dict)
  # Lowercase the document
  doc_dict = doc_dict.lower()
  # Remove punctuations
  doc_dict = re.sub(r'[%s]' % re.escape(string.punctuation), ' ', doc_dict)
  # Remove the numbers
  # Could be a problem in the future
  doc_dict = re.sub(r'[0-9]', '', doc_dict)
  # Remove the doubled space
  doc_dict = re.sub(r'\s{2,}', ' ', doc_dict)
  # Remove the stop words from each word in a document
  filtered_doc_dict = ' '.join([word for word in doc.split(" ") if not word in stop_words])
  d["contents"] = filtered_doc_dict

# Instantiate a TfidfVectorizer object
# We use TFidfVectorizer to downweight the words that appear more frequently across documents
vectorizer = TfidfVectorizer()
# Fit the data and transform it as a vector
X = vectorizer.fit_transform(clean_doclist)
# Convert the X as a transposed matrix
X = X.T.toarray()
# Create a DataFrame and set the vocabulary as the index
df = pd.DataFrame(X, index=vectorizer.get_feature_names())

#print(df)
get_similar_articles("pcr", df)
```
